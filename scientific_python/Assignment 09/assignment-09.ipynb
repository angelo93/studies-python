## Requirement 04

Numpy arrays are at face value very similar to Python Lists. In fact, it is failry easy to interchange the two if doing simple operations. Additionally, they are also very syntactically similar and both can be manipulated in fairly similar ways. The biggest difference arises once someone takes a peek under the hood, so to say. Numpy arrays are more akin to C style arrays in that only the same type may exist in the array and they are primitive types to the fullest extent of the definition. Python Lists on the other hand are arrays of pointers to an object stored in memory where additional information about the element is stored. This is how Python Lists can have the added bonus of storing any kind of type at once in the same list. 

The biggest negative the Python Lists incur is their speed and ability to do more complex operations. This is due to how the Lists are set up in memory. On the other hand, their flexibility is a tremendous asset that allows them to act as something more than a list. In sense they can act like structures to a certain degree. Numpy arrays, as previously stated, follow more static typing norms which means they require less memory space and are able to have operations done on the array entries at a much faster rate. As with most things in the coding world, every tool has their use, it is up to the coder to know which tool is best for the job and, more importantly, how the tool actually works and how to use it.

## Requirement 05

Variable typing in Python is Dynamic, in that while a variable may be declared and initialized with a certain type, the typing can change on the fly during runtime. This is mainly due to the fact that Python is not a compiled language in the traditional sense like the C language. Instead, Python has an interpreter that acts as an intermediary during runtime that converts Python code into machine language. This allows Python to be able to do nifty tricks, like changing the typing of a variable. Languages like C rely on consitency and compiled code to work and interact in a certain way. In fact, depending on the IDE and compiler, C won't even let you compile correctly if the typing of a variable is changed in your code.

As previously stated, there are pros and cons to both of these tools. Python's flexibility means the learning curve and level of entry is much lower than other languages. Javascript also "suffers" from dynmaic typing and so typescript was born and the cult like following of using triple equals because double equals doesn't check if two variables are of the same type before comparing. Funnily enough, it actually does, the very first thing double equals does is check types and then coerce the typing of each variable with a small and strict algorithm. I digress though, again, languages and language features are tools and like any tool it can be mishandled and used in the wrong situation unless proper knowledge and wisdom of the tool is accrued and practices are adhered to.

## Requirement 19

A NumPy universal function (ufunc) is a a special vectorized operation that is designed to execute at the lower compiled layer of NumPy. They work by applying the operation to each element of an array and doing so quickly because they do not have to worry about as many of the hinderances of being at a higher level of the Python Interpretation layer chain. This is especially helpful when doing repeated operations on large size arrays as it essentially eliminates type checking and operation lookups.

## Requirement 23

```python
# Demonstrate Numpy's...
# 1. Sum, min & max functions
# 2. Time with %timeit
# 3. Compare with Python's sum, min & max

numpy_array = np.random.randint(1, 100, size=10)

print(f'Original Array: {numpy_array}')
print(f'Sum of Array: {np.sum(numpy_array)}')
print(f'Min of Array: {np.min(numpy_array)}')
print(f'Max of Array: {np.max(numpy_array)}')

print('\nSum Method Comparison Python Vs NumPy')
%timeit sum(numpy_array)
%timeit np.sum(numpy_array)
print('\nMin Method Comparison Python Vs NumPy')
%timeit min(numpy_array)
%timeit np.min(numpy_array)
print('\nMax Method Comparison Python Vs NumPy')
%timeit max(numpy_array)
%timeit np.max(numpy_array)
```

```python
# Use boradcasting and Matplotlib to plot produce a plot of a two-dimensional array

array_x = np.linspace(0, 19, 100)
array_y = np.linspace(0, 100, 100)[:, np.newaxis]

array_z = np.sin(array_x) ** 10 + np.cos(10 + array_y * array_x) * np.cos(array_x)

%matplotlib inline
import matplotlib.pyplot as plt

plt.imshow(array_z, origin='lower', extent=[0, 5, 0, 5],
           cmap='viridis')
plt.colorbar();
```
