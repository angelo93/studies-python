## Requirement 04

```python
# Use the Matplotlib object-oriented interface to display the sine, cosine,
# tangent, and arctangent of x, each in its own plot

x = np.linspace(0, 10, 100)
fig, ax = plt.subplots(4, figsize=(12,12))

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
ax[2].plot(x, np.tan(x))
ax[3].plot(x, np.arctan(x));
```

## Requirement 07

```python
# Plot the hyperbolic tangent (tanh()) function using six different color
# formats. Modify x slightly in each plot so the lines are distinct

x = np.linspace(0, 10, 100)

plt.plot(x, np.tanh(x - 0), color='blue')        
plt.plot(x, np.tanh(x - 1), color='BlueViolet')  
plt.plot(x, np.tanh(x - 2), color='CadetBlue')  
plt.plot(x, np.tanh(x - 3), color='DarkOrange')  
plt.plot(x, np.tanh(x - 4), color='DarkRed') 
plt.plot(x, np.tanh(x - 5), color='DarkSlateGray');

plt.xlim(0, 10);
```

## Requirement 08

```python
# Using the same x as above, plot the cosine (cos()) using 4 different color
# formats and 4 different line types. Modify x slightly in each plot so the lines
# are distinct

x = np.linspace(0, 10, 100)

plt.plot(x, np.cos(x - 0), color='blue', linestyle='-')        
plt.plot(x, np.cos(x - 1), color='BlueViolet', linestyle='--')  
plt.plot(x, np.cos(x - 2), color='CadetBlue', linestyle='-.')  
plt.plot(x, np.cos(x - 3), color='DarkRed', linestyle=':') 

plt.xlim(0, 10);
```

## Requirement 09

```python
# Demonstrate limiting the x and y axes values in a plot

x = np.linspace(0, 10, 100)

plt.plot(x, np.cos(x - 0), color='blue', linestyle='-')        
plt.plot(x, np.cos(x - 1), color='BlueViolet', linestyle='--')  
plt.plot(x, np.cos(x - 2), color='CadetBlue', linestyle='-.')  
plt.plot(x, np.cos(x - 3), color='DarkRed', linestyle=':') 

plt.xlim(0, 10)
plt.ylim(-1, 1);
```

## Requirement 10

```python
# Demonstrate a plot with an equal aspect ratio

x = np.linspace(0, 10, 100)

plt.plot(x, np.cos(x - 0), color='blue', linestyle='-')        
plt.plot(x, np.cos(x - 1), color='BlueViolet', linestyle='--')  
plt.plot(x, np.cos(x - 2), color='CadetBlue', linestyle='-.')  
plt.plot(x, np.cos(x - 3), color='DarkRed', linestyle=':') 

plt.axis('equal');
```

## Requirement 11

```python
# Demonstrate the use of axis labels and a plot legend

x = np.linspace(0, 10, 100)

plt.plot(x, np.cos(x - 0), color='blue', linestyle='-')        
plt.plot(x, np.cos(x - 1), color='BlueViolet', linestyle='--')  
plt.plot(x, np.cos(x - 2), color='CadetBlue', linestyle='-.')  
plt.plot(x, np.cos(x - 3), color='DarkRed', linestyle=':') 

plt.axis('tight')

plt.title("Cosine 4 Ways")
plt.xlabel("x")
plt.ylabel("cos(x)");
```

## Requirement 12

```python
# Use RandomState() and plot() to demonstrate 11 different plot marker
# types. Also, display the markers in a legend

rng = np.random.RandomState(42)

for marker in ['3', '.', ',', '4', '+', 'h', '^', '<', '>', '1', '2']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
# end for

plt.legend(numpoints=1)
plt.xlim(0, 1.5);
```

## Requirement 17

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
# end def f

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='PuOr', alpha=0.5)
plt.colorbar();
```

## Requirement 19

```python
plt.style.use('classic')

fig, ax = plt.subplots()

lines = []
styles = ['-', '--', '-.', ':']
x = np.linspace(0, 10, 1000)

for i in range(4):
    lines += ax.plot(x, np.sin(x - i * np.pi / 2),
                     styles[i], color='#7aa329', linewidth=2)
ax.axis('equal')
ax.set_xlim(0,10)
ax.set_ylim(-2, 2)

ax.legend(lines[:2], ['line A', 'line B'],
          loc='upper left', frameon=True)

from matplotlib.legend import Legend
leg = Legend(ax, lines[2:], ['line C', 'line D'],
             loc='lower left', frameon=True)
ax.add_artist(leg);
```
